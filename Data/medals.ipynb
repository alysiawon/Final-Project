```python
# Import Dependencies
import pandas as pd
import os as os
import numpy as np
```

### 120 Years of Olympic Data

```python
# Open 120 Years of Olympic Data File
olympic_athletes_file_to_load = os.path.join("Resources","120 Years Of Olympic Data", "athlete_events.csv")
olympic_athletes_df = pd.read_csv(olympic_athletes_file_to_load)
olympic_athletes_df.head()
```

```python
# Remove Bronze & Silver
gold_olympic_athletes_df = olympic_athletes_df[olympic_athletes_df.Medal == "Gold"]
gold_olympic_athletes_df.head()
```

```python
# Drop Columns
gold_olympic_athletes_df.drop(["ID","Team","Games","Season","City"], axis=1, inplace=True)
gold_olympic_athletes_df.head()
```

### 2020 Olympics Data

```python
# Open 2020 Olympic Athlete Data File
current_olympic_athletes_file_to_load = os.path.join("Resources","2021 Olympics", "athletes.csv")
current_olympic_athletes_df = pd.read_csv(current_olympic_athletes_file_to_load)
current_olympic_athletes_df.head()
```

```python
# Drop Columns
current_olympic_athletes_df.drop(["short_name","birth_place","birth_country","residence_place","residence_country","url"], axis=1, inplace=True)
current_olympic_athletes_df.head()
```

```python
# Open 2020 Olympic Medal Data File
current_olympic_medals_file_to_load = os.path.join("Resources","2021 Olympics", "medals.csv")
current_olympic_medals_df = pd.read_csv(current_olympic_medals_file_to_load)
current_olympic_medals_df.head()
```

```python
# Remove Bronze & Silver
gold_current_olympic_athletes_df = current_olympic_medals_df[current_olympic_medals_df.medal_type == "Gold Medal"]
gold_current_olympic_athletes_df.head()
```

```python
# Drop Columns
gold_current_olympic_athletes_df.drop(["medal_code","medal_date","athlete_short_name","athlete_link"], axis=1, inplace=True)
gold_current_olympic_athletes_df.head()
```

### GDP

```python
# Open GDP Data File
gdp_file_to_load = os.path.join("World GDP", "gdp.csv")
gdp_df = pd.read_csv(gdp_file_to_load, error_bad_lines=False)
gdp_df
```

```python
# Drop Columns
gdp_df.drop(["Indicator Name","Indicator Code"], axis=1, inplace=True)
```

```python
#Transpose Data
gdp_df = pd.melt(gdp_df, id_vars=["Country Name", "Country Code"], 
                  var_name="Year", value_name="Value")
```

```python
#Change Data Type
gdp_df.Year = gdp_df.Year.astype(np.int64)
```

```python
# Rename Headers
gdp_df = gdp_df.rename(columns={"Value": "GDP"})
gdp_df
```

```python
# Drop Empty Rows
gdp_df.dropna()
```

```python
# Reorder Columns
gdp_df = gdp_df.reindex(['Year', 'Country Name','Country Code','GDP'], axis=1)
gdp_df
```

### Population

```python
# Open Population Data File
population_file_to_load = os.path.join("World Population", "population.csv")
population_df = pd.read_csv(population_file_to_load)
population_df
```

```python
# Drop Columns
population_df.drop(["Yearly % Change", "Yearly Change", "Migrants (net)", "Median Age", 
                    "Fertility Rate","Density (P/Km)","Urban Pop %","Urban Population", 
                    "Country's Share of World Pop %", "Country Global Rank", "World Population"], axis=1, inplace=True)
population_df.head()
```

```python
# Rename Headers
population_df = population_df.rename(columns={"Country": "Country Name"})
population_df
```

### Merge Dataframe

```python
# Merge Dataframe
country_df = pd.merge(gdp_df, population_df, on=['Year','Country Name'], how="left")
country_df
```

```python
# Open Athletes Data File
athletes_file_to_load = os.path.join("athletes.csv")
athletes_df = pd.read_csv(athletes_file_to_load)
athletes_df
```

```python
# Rename Headers
athletes_df = athletes_df.rename(columns={"NOC": "Code"})
```

```python
# Group By Year and Country
athletes = athletes_df.groupby(["Year", "Code"]).size().reset_index(name="Gold Medals")
```

```python
# New Dataframe
athlete_df = pd.DataFrame(athletes)
athlete_df
```

```python
# Rename Headers
athlete_df = athlete_df.rename(columns={"Code": "Country Code"})
athlete_df
```

### Country Data

```python
# Open Country Data File
country_file_to_load = os.path.join("country.csv")
country_df = pd.read_csv(country_file_to_load)
```

```python
# Reorder Columns
country_df = country_df.reindex(['Year', 'Country Code','Country Name','GDP','Population'], axis=1)
country_df
```

### New Dataframe: Medals Data

```python
# Merge DataFrame
medals_df = pd.merge(country_df,athlete_df, on=['Year','Country Code'], how="left")
medals_df
```

```python
# Replace NaN with 0
medals_df['Gold Medals'] = medals_df['Gold Medals'].replace(np.nan, 0)

# Reorder Columns
medals_df = medals_df.reindex(['Country Code', 'Year','Country Name', "GDP", "Population", "Gold Medals"], axis=1)
```

### Coordinates

```python
# Open Country Code Data File
coordinates_file_to_load = os.path.join("Resources","World Coordinates","coordinates.csv")
coordinates_df = pd.read_csv(coordinates_file_to_load)
coordinates_df
```

```python
# Drop Columns
coordinates_df.drop(["usa_state_code","usa_state_latitude","usa_state_longitude","usa_state"], axis=1, inplace=True)
```

```python
# Open Coordinate Data File
country_code_file_to_load = os.path.join("Resources","World Coordinates","country_codes.csv")
country_code_df = pd.read_csv(country_code_file_to_load)
country_code_df
```

```python
# Drop Columns
country_code_df.drop(["Numeric code","ISO 3166-2"], axis=1, inplace=True)

# Rename Headers
country_code_df = country_code_df.rename(columns={"English short name lower case": "country", "Alpha-2 code":"country_code"})

# Reorder Columns
country_code_df = country_code_df.reindex(['country_code', 'country','Alpha-3 code'], axis=1)
country_code_df
```

```python
# Merge DataFrame
coordinate_df = pd.merge(coordinates_df,country_code_df, on=['country_code'], how="left")
coordinate_df
```

```python
# Drop Columns
coordinate_df.drop(["country_x","country_y","country_code"], axis=1, inplace=True)

# Rename Headers
coordinate_df = coordinate_df.rename(columns={"Alpha-3 code": "Country Code", "latitude":"Latitude","longitude":"Longitude"})

# Reorder Columns
coordinate_df = coordinate_df.reindex(['Country Code', 'Latitude','Longitude'], axis=1)
coordinate_df
```

```python
# Merge DataFrame
medals_coordinates_df = pd.merge(coordinate_df,medals_df, on=['Country Code'], how="right")
medals_coordinates_df
```

```python
# Replace NaN with 0
medals_coordinates_df = medals_coordinates_df.replace(np.nan, 0)
medals_coordinates_df
```

```python
# Reorder Columns
medals_coordinates_df = medals_coordinates_df.reindex(['Year', 'Country Name','Country Code','Latitude','Longitude','GDP','Population', 'Gold Medals'], axis=1)
medals_coordinates_df
```

```python
# Rename Headers
medals_coordinates_df = medals_coordinates_df.rename(columns={"Year": "year", "Country Name":"country_name","Country Code":"country_code", "GDP": "gdp", "Population":"population", "Gold Medals": "gold_medals", "Longitude":"longitude", "Latitude": "latitude"})
medals_coordinates_df
```

```python
medals_coordinates_df.dtypes
```

```python
# Convert Datatypes
medals_coordinates_df['gold_medals'] = medals_coordinates_df['gold_medals'].astype(np.int64)
medals_coordinates_df['gdp'] = medals_coordinates_df['gdp'].astype(np.int64)
medals_coordinates_df['population'] = medals_coordinates_df['population'].astype(np.int64)
medals_coordinates_df.dtypes
```

```python
# Export CSV
medals_coordinates_df.to_csv('medals.csv', index=False)
```
