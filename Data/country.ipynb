```python
# Import Dependencies
import pandas as pd
import os as os
import numpy as np
```

### GDP

```python
# Open GDP Data File
gdp_file_to_load = os.path.join("World GDP", "gdp.csv")
gdp_df = pd.read_csv(gdp_file_to_load, error_bad_lines=False)
gdp_df
```

```python
# Drop Columns
gdp_df.drop(["Indicator Name","Indicator Code"], axis=1, inplace=True)
```

```python
#Transpose Data
gdp_df = pd.melt(gdp_df, id_vars=["Country Name", "Country Code"], 
                  var_name="Year", value_name="Value")
```

```python
#Change Data Type
gdp_df.Year = gdp_df.Year.astype(np.int64)
```

```python
# Rename Headers
gdp_df = gdp_df.rename(columns={"Value": "GDP"})
gdp_df
```

```python
# Drop Empty Rows
gdp_df.dropna()
```

```python
# Reorder Columns
gdp_df = gdp_df.reindex(['Year', 'Country Name','Country Code','GDP'], axis=1)
gdp_df
```

### Population

```python
# Open Population Data File
population_file_to_load = os.path.join("World Population", "population.csv")
population_df = pd.read_csv(population_file_to_load)
population_df
```

```python
# Drop Columns
population_df.drop(["Yearly % Change", "Yearly Change", "Migrants (net)", "Median Age", 
                    "Fertility Rate","Density (P/Km²)","Urban Pop %","Urban Population", 
                    "Country's Share of World Pop %", "Country Global Rank", "World Population"], axis=1, inplace=True)
population_df.head()
```

```python
# Rename Headers
population_df = population_df.rename(columns={"Country": "Country Name"})
population_df
```

### Merge Dataframe

```python
# Merge Dataframe
country_df = pd.merge(gdp_df, population_df, on=['Year','Country Name'], how="left")
country_df
```

```python
# Export CSV
country_df.to_csv('country.csv', index=False)
```
