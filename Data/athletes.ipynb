```python
# Import Dependencies
import pandas as pd
import os as os
from datetime import datetime, date
```

## 120 Years of Olympic Data

```python
# Open 120 Years of Olympic Data File
olympic_athletes_file_to_load = os.path.join("Resources","120 Years Of Olympic Data", "athlete_events.csv")
olympic_athletes_df = pd.read_csv(olympic_athletes_file_to_load)
olympic_athletes_df.head()
```

```python
# Remove Bronze & Silver
gold_olympic_athletes_df = olympic_athletes_df[olympic_athletes_df.Medal == "Gold"]
gold_olympic_athletes_df.head()
```

```python
# Drop Columns
gold_olympic_athletes_df.drop(["ID","Team","Games","Season","City"], axis=1, inplace=True)
gold_olympic_athletes_df.head()
```

### 2020 Olympics Data

```python
# Open 2020 Olympic Athlete Data File
current_olympic_athletes_file_to_load = os.path.join("Resources","2021 Olympics", "athletes.csv")
current_olympic_athletes_df = pd.read_csv(current_olympic_athletes_file_to_load)
current_olympic_athletes_df.head()
```

```python
# Drop Columns
current_olympic_athletes_df.drop(["short_name","birth_place","birth_country","residence_place","residence_country","url","discipline_code"], axis=1, inplace=True)
current_olympic_athletes_df.head()
```

```python
# Clean Gender
current_olympic_athletes_df['gender'].replace(["Female","Male"],["F","M"],inplace=True)
current_olympic_athletes_df
```

```python
# Clean Age
def age(born):
    born = datetime.strptime(born, "%Y-%m-%d").date
    today = date.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))
  
current_olympic_athletes_df['Age'] = current_olympic_athletes_df['birth_date'].apply(age)
```

```python
# Clean Height

# Separate m and ft, drop extra rows
current_olympic_athletes_df[["Height", "Height (feet)"]] = current_olympic_athletes_df['height_m/ft'].str.split("/", expand=True)
current_olympic_athletes_df.drop(["height_m/ft","Height (feet)"], axis=1, inplace=True)

# Change to Float
current_olympic_athletes_df["Height"] = current_olympic_athletes_df.Height.astype(float)

# Change m to cm
current_olympic_athletes_df["Height"] = current_olympic_athletes_df["Height"] * 100

current_olympic_athletes_df
```

```python
# Rename Headers
current_olympic_athletes_df = current_olympic_athletes_df.rename(columns={"name": "Name", "gender": "Sex", "birth_date": "Age", "country": "Country","country_code": "NOC", "event": "Event", "discipline": "Sport"})
current_olympic_athletes_df
```

### Medals

```python
# Open 2020 Olympic Medal Data File
current_olympic_medals_file_to_load = os.path.join("Resources","2021 Olympics", "medals.csv")
current_olympic_medals_df = pd.read_csv(current_olympic_medals_file_to_load)
current_olympic_medals_df.head()
```

```python
# Remove Bronze & Silver
gold_current_olympic_athletes_df = current_olympic_medals_df[current_olympic_medals_df.medal_type == "Gold Medal"]
gold_current_olympic_athletes_df.head()
```

```python
# Drop Columns
gold_current_olympic_athletes_df.drop(["medal_code","medal_date","athlete_short_name","athlete_link", "discipline_code", "country"], axis=1, inplace=True)
gold_current_olympic_athletes_df.head()
```

```python
# Clean Gender
gold_current_olympic_athletes_df['athlete_sex'].replace(["W","M","X"],["F","M","X"],inplace=True)
gold_current_olympic_athletes_df
```

```python
# Rename Headers
gold_current_olympic_athletes_df = gold_current_olympic_athletes_df.rename(columns={"medal_type": "Medal", "athlete_name": "Name", "athlete_sex": "Sex", "country_code": "NOC", "event": "Event", "discipline": "Sport"})
gold_current_olympic_athletes_df
```
