```python
# Import Dependencies
import pandas as pd
import os as os
```

### GDP

```python
# Open GDP Data File
gdp_file_to_load = os.path.join("Resources","World GDP", "gdp_data.csv")
gdp_df = pd.read_csv(gdp_file_to_load)
gdp_df
```

```python
# Drop Columns
gdp_df.drop(["GDP-Growth","GDP-Per-Capita"], axis=1, inplace=True)
gdp_df
```

```python
# Rearrange Columns
gdp_df.reindex(['Year', 'Country','Code','GDP'], axis=1)
```

### Population

```python
# Open Population Data File
population_file_to_load = os.path.join("Resources","World Population", "population.csv")
population_df = pd.read_csv(population_file_to_load)
population_df
```

```python
# Drop Columns
population_df.drop(["Yearly % Change", "Yearly Change", "Migrants (net)", "Median Age", 
                    "Fertility Rate","Density (P/Km²)","Urban Pop %","Urban Population", 
                    "Country's Share of World Pop %", "Country Global Rank"], axis=1, inplace=True)
population_df
```

### Merge Dataframe

```python
# Merge Dataframe
pd.merge(gdp_df, population_df, on=['Year','Country'])
```
